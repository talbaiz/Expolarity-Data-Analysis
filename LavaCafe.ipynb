## Visualization

```python
plt.figure(figsize=[15,6])
sns.set_theme(style="whitegrid")
plt.xlabel("shop_name",fontsize=18 , family = 'sans' , weight='bold' , color = '#808080')
plt.ylabel("rating",fontsize=18 , family = 'sans' , weight='bold' , color = '#808080' )
plt.title('The Most Rating Of Coffee Shop',fontsize=22 , family = 'sans' , weight='bold' , color = '#808080')
sns.barplot(x="shop_name", y="rating", data=df1.iloc[7:22,:]) # plotting From row 7 to 20 and plots All columns
plt.xticks(rotation=90, fontsize = 12);
plt.yticks(rotation=0, fontsize = 12);
```

```python
plt.figure(figsize=[20,8])
sns.set_theme(style="whitegrid")
plt.xlabel("shop_type",fontsize=28 , family = 'sans' , weight='bold' , color = '#808080')
plt.ylabel("price",fontsize=28 , family = 'sans' , weight='bold' , color = '#808080' )
plt.title("The Different Price Between Cafe and Coffee Shops", fontsize=28  , family = 'sans' , weight='bold' , color = '#808080')
sns.barplot(x="shop_type", y="price", data=df1.iloc[6:30,:] )
plt.xticks(rotation=0 ,  fontsize = 22 );
```
